```
import pandas as pd
import operator
from sklearn.model_selection import train_test_split
import random
import numpy as np
from sklearn.svm import OneClassSVM


train_filename = "/Users/junaidrasool/Desktop/comp90073_2/A2_1/NMF/train1_NMF.csv"
train_file = open(train_filename)
df_train = pd.read_csv(train_file,index_col=0)
df_train

test_filename = "/Users/junaidrasool/Desktop/comp90073_2/A2_1/NMF/test1_NMF.csv"
test_file = open(test_filename)
df_test = pd.read_csv(test_file,index_col=0)
df_test

train_sample = df_train.sample(frac =.005) 
train_sample

clf = OneClassSVM(gamma='auto').fit(train_sample)

df_test1_NMF=df_test.iloc[0:40000]
df_test1_NMF=df_test1_NMF.reset_index(drop=True)

df_test2_NMF=df_test.iloc[40000:80000]
df_test2_NMF=df_test2_NMF.reset_index(drop=True)

df_test3_NMF=df_test.iloc[80000:120000]
df_test3_NMF=df_test3_NMF.reset_index(drop=True)

df_test4_NMF=df_test.iloc[120000:160000]
df_test4_NMF=df_test4_NMF.reset_index(drop=True)

df_test5_NMF=df_test.iloc[160000:200000]
df_test5_NMF=df_test5_NMF.reset_index(drop=True)

df_test6_NMF=df_test.iloc[200000:240000]
df_test6_NMF=df_test6_NMF.reset_index(drop=True)

predict_test1=clf.predict(df_test1_NMF)

predict_test2=clf.predict(df_test2_NMF)

predict_test3=clf.predict(df_test3_NMF)

predict_test4=clf.predict(df_test4_NMF)

predict_test5=clf.predict(df_test5_NMF)

predict_test6=clf.predict(df_test6_NMF)

predict_test1

predict_test2

predict_test3

predict_test4

predict_test5

predict_test6

df_test1_NMF_label = pd.DataFrame(predict_test1,columns= ['label'])
df_test1_NMF_label

df_test2_NMF_label = pd.DataFrame(predict_test2,columns= ['label'])
df_test2_NMF_label

df_test3_NMF_label = pd.DataFrame(predict_test3,columns= ['label'])
df_test3_NMF_label

df_test4_NMF_label = pd.DataFrame(predict_test4,columns= ['label'])
df_test4_NMF_label

df_test5_NMF_label = pd.DataFrame(predict_test5,columns= ['label'])
df_test5_NMF_label

df_test6_NMF_label = pd.DataFrame(predict_test6,columns= ['label'])
df_test6_NMF_label

df_test1_NMF['label'] = df_test1_NMF_label

df_test2_NMF['label'] = df_test2_NMF_label

df_test3_NMF['label'] = df_test3_NMF_label

df_test4_NMF['label'] = df_test4_NMF_label

df_test5_NMF['label'] = df_test5_NMF_label

df_test6_NMF['label'] = df_test6_NMF_label

df_test1_NMF

df_test2_NMF

df_test3_NMF

df_test4_NMF

df_test5_NMF

df_test6_NMF

vertical_stack = pd.concat([df_test1_NMF, df_test2_NMF], axis=0)
vertical_stack=vertical_stack.reset_index(drop=True)

vertical_stack = pd.concat([vertical_stack, df_test3_NMF], axis=0)
vertical_stack=vertical_stack.reset_index(drop=True)

vertical_stack = pd.concat([vertical_stack, df_test4_NMF], axis=0)
vertical_stack=vertical_stack.reset_index(drop=True)

vertical_stack = pd.concat([vertical_stack, df_test5_NMF], axis=0)
vertical_stack=vertical_stack.reset_index(drop=True)

vertical_stack = pd.concat([vertical_stack, df_test6_NMF], axis=0)
vertical_stack=vertical_stack.reset_index(drop=True)

vertical_stack

vertical_stack.nunique()


vertical_stack.to_csv('test_pred_NMF_OCSVM.csv')

```
