```
import pandas as pd
from sklearn.neural_network import MLPClassifier
import random
import numpy as np
from sklearn.metrics import confusion_matrix

train_filename = "/Users/junaidrasool/Desktop/comp90073_2/A2_2/train_2.csv"
train_file = open(train_filename)
df_train = pd.read_csv(train_file,index_col=0)
df_train

valid_filename = "/Users/junaidrasool/Desktop/comp90073_2/A2_2/valid_2.csv"
valid_file = open(valid_filename)
df_valid = pd.read_csv(valid_file,index_col=0)
df_valid

test_filename = "/Users/junaidrasool/Desktop/comp90073_2/A2_2/test_2.csv"
test_file = open(test_filename)
df_test = pd.read_csv(test_file,index_col=0)
df_test_final

vertical_stack = pd.concat([df_train, df_valid], axis=0)
vertical_stack=vertical_stack.reset_index(drop=True)

vertical_stack

df_train = vertical_stack.select_dtypes(exclude=['object'])
df_train

df_test = df_test.select_dtypes(exclude=['object'])
df_test

df_train = df_train.replace(np.nan, 0, regex=True)



df_test = df_test.replace(np.nan, 0, regex=True)

df_train_features= df_train.drop(['label'],axis=1)
df_train_features

df_train_label=df_train['label']
df_train_label

df_test_features= df_test.drop(['label'],axis=1)
df_test_features

df_test_label=df_test['label']
df_test_label



clf = MLPClassifier(hidden_layer_sizes=(50,50,50), max_iter=10,activation = 'relu',solver='adam',random_state=1,batch_size=1000)

clf.fit(df_train_features, df_train_label)

test_pred = clf.predict(df_test_features)

test_pred = pd.DataFrame(test_pred,columns= ['pred_label'])

cm = confusion_matrix(df_test_label, test_pred)

cm

TN= 508984
FP=4757
FN=58545
TP=2096

Accuracy = (TP + TN)/(TP + TN + FP + FN)
Accuracy

precision = TP / (TP + FP)
precision

recall = TP / (TP + FN)
recall

f1_score = 2 * precision * recall / (precision + recall)
f1_score

df_test['pred_label']= test_pred

df_test

test_filename = "/Users/junaidrasool/Desktop/comp90073_2/A2_2/test_2.csv"
test_file = open(test_filename)
df_test_final = pd.read_csv(test_file,index_col=0)
df_test_final

df_test_final['pred_label']= test_pred

df_test_final

df_test_task2=df_test_final.loc[df_test_final['pred_label'] == -1]

df_test_task2

df_test_task2.to_csv('task2.csv')


```
